```python
# Problem Statement 1:Smart Parking Lot Management System

# Function to manage parking lot
def parking_system(capacity, logs):

    parked = 0
    peak = 0

    # Loop through vehicle logs
    for log in logs:

        if log == "IN":
            parked += 1

        elif log == "OUT":
            parked -= 1

        # Track peak usage
        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    # Check parking status
    if parked > capacity:
        print("Parking Status: Capacity Exceeded")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# Function call
parking_system(capacity, vehicle_logs)
```

```python
# Problem Statement 2:Online Food Delivery Time Estimator
# Function to estimate delivery time
def delivery_time(distance, traffic, weather):

    # Base time calculation (5 minutes per km)
    time = distance * 5

    # Traffic condition adjustment
    if traffic.lower() == "high":
        time += 10
    elif traffic.lower() == "medium":
        time += 5

    # Weather condition adjustment
    if weather.lower() == "rainy":
        time += 5
    elif weather.lower() == "storm":
        time += 10

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

# Function call
delivery_time(distance, traffic, weather)
```

```python
# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
# Function to analyze seat occupancy
def seat_analyzer(total_seats, booked_seats):

    booked_count = 0

    # Loop to count booked seats
    for seat in booked_seats:
        if seat == 1:
            booked_count += 1

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", occupancy, "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

    # Suggest additional shows
    if occupancy >= 90:
        print("Suggestion: Open Additional Show")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # represents 150 booked seats

# Function call
seat_analyzer(total_seats, booked_seats)
```

```python
# Problem Statement 4:Cloud Server Load Classification System
# Function to classify server load
def server_load(cpu_readings):

    total = 0

    # Loop to calculate total CPU usage
    for cpu in cpu_readings:
        total += cpu

    # Calculate average CPU load
    avg = total / len(cpu_readings)

    print("Average CPU Load:", avg, "%")

    # Classify server status
    if avg < 50:
        print("Server Status: Normal")

    elif avg <= 80:
        print("Server Status: Warning")

    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

# Function call
server_load(cpu_readings)
```

```python
#  Problem Statement 5:Smart Classroom Resource Usage Monitor
# Function to monitor resource usage
def resource_monitor(resources):

    overused = []

    # Loop through dictionary
    for resource, hours in resources.items():

        # Check if usage is more than 8 hours
        if hours > 8:
            overused.append(resource)

    # Display overused resources
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Function call
resource_monitor(resource_usage)
```

```python
# Problem Statement 6: Online Event Registration Capacity Controller
# Function to manage event registrations
def event_registration(capacity, registrations):

    confirmed = 0
    waitlist = 0

    # Loop to process registrations
    for i in range(registrations):

        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)

    # Registration status
    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
capacity = 100
registrations = 105

# Function call
event_registration(capacity, registrations)
```
